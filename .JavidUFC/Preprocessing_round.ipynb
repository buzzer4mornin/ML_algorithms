# Balancing dataset

#### It was observed that the winner from 1993 until 2009 (1292 fights) are red fighters only. So, we will balance that data in accordance with the proportion after 2010 (59% and 41%). 

```python
#Converting date column to date format
df['date'] = pd.to_datetime(df.date, format='%Y-%m-%d')
```

```python
# Subset data by date and drop from main df
df_balance = df[(df['date'] >= '1994-03-11') &
          (df['date'] <= '2009-12-12')]
df.drop(df_balance.index, inplace=True)
```

```python
df_balance
```

```python
# Check number of values
df_balance['Winner'].value_counts()
```

```python
# Change 504 Red to Blue in order to correspond to overall proportion (59% and 41%)
sample = df_balance[df_balance['Winner']=='Red'].sample(n=504)
df_balance.drop(sample.index, inplace=True)
```

```python
# Rename columns for changed winner color
rename_cols = {}
for column in sample.columns:
    if column.startswith('B_'):
        rename_cols['B_' + column.split('B_', 1)[1]] = 'R_' + column.split('B_', 1)[1]
        rename_cols['R_' + column.split('B_', 1)[1]] = 'B_' + column.split('B_', 1)[1]

sample.rename(rename_cols, axis='columns', inplace=True)
sample.replace({'Winner': 'Red'}, 'Blue', inplace=True)
```

```python
# Concatenate sample with the df_falance to get the initially created subset (1229)
df_balance_final = pd.concat([df_balance, sample]).sort_index()
```

```python
# 725 Red and 504 Blue values created
df_balance_final['Winner'].value_counts()
```

```python
# Concatenate with the main df file
df = pd.concat([df, df_balance_final]).sort_index()
```
